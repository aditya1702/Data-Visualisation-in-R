## Diving into ggplot

There are 2 variables in the above data:

1. displ - a car's engine size, in litres
2. hwy - a car’s fuel efficiency on the highway, in miles per gallon (mpg)

We want to analyse the relationship between these 2 variables. We create our first plot in ggplot2

````r
options(repr.plot.width = 10, repr.plot.height = 5)
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy))
````

1. The plot shows a negative relationship between engine size (displ) and fuel efficiency (hwy). In other words, cars with big engines use more fuel. 
2. With ggplot2, you begin a plot with the function ggplot(). ggplot() creates a coordinate system that you can add layers to. The first argument of ggplot() is the dataset to use in the graph. So ggplot(data = mpg) creates an empty graph, but it’s not very interesting so I’m not going to show it here. 
3. You complete your graph by adding one or more layers to ggplot(). The function geom_point() adds a layer of points to your plot, which creates a scatterplot. ggplot2 comes with many geom functions that each add a different type of layer to a plot.
4. Finally, every geom() function in ggplot has an attibute aes() which maps the variables in the data with visual properties of a graph.

## Create a basic template

The reason ggplot is used so much is because any visualisation can be created using a basic template. Once this template is created, different graphs can be plotted using this template.

```
ggplot(data = <DATA>) + 
<GEOM_FUNCTION>(mapping = aes(<MAPPINGS>))
```

The above template is how all ggplot code snippets will look like.

## Aesthetic Mappings

What if we want to add some more information in our graphs. Say, we are analysing relationship between 2 variables in our dataset and we also want to add a third variable. How can we do it? This is where aesthetic mappings come into play. We use the previous example.

````r
options(repr.plot.width = 10, repr.plot.height = 5)
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy, color = class))
````

To map an aesthetic to a variable, associate the name of the aesthetic to the name of the variable inside aes(). ggplot2 will automatically assign a unique level of the aesthetic (here a unique color) to each unique value of the variable, a process known as **scaling**. ggplot2 will also add a legend that explains which levels correspond to which values.

In the above example, we mapped class to the color aesthetic, but we could have mapped class to the size aesthetic in the same way. In this case, the exact size of each point would reveal its class affiliation.

````r
options(repr.plot.width = 10, repr.plot.height = 5)
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy, size = class))
````

There are other aesthetics too - **alpha and shape**.

````r
options(repr.plot.width = 10, repr.plot.height = 5)
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy, alpha = class))
````

````r
options(repr.plot.width = 10, repr.plot.height = 5)
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy, shape = class))
````

The above examples showed how you can use different aesthetics to map variables. You can also change the aesthetics of your plots

````r
options(repr.plot.width = 10, repr.plot.height = 5)
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy), color = "red")
````

## Facet Plots

One way to add additional variables is with aesthetics. Another way, particularly useful for categorical variables, is to split your plot into facets, subplots that each display one subset of the data. This makes it easy to compare the data - by dividing it into multiple subplots

````r
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy)) + 
facet_wrap(~ class, nrow = 2)
````

Instead of a single variable, we can also facet onto 2 variables.

````r
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy)) + 
facet_grid(drv ~ cyl)
````

## Using Different Geoms

Earlier, we wanted to see the relationship between displ and hwy and we chose to use a scatter plot or geom_point() function. However, we can also choose from other different geom() functions offered by ggplot

````r
ggplot(data = mpg) + 
geom_smooth(mapping = aes(x = displ, y = hwy))
````

Again, we can add aesthetic mappings for a third variable just like we did for the point plots

````r
ggplot(data = mpg) + 
geom_smooth(mapping = aes(x = displ, y = hwy, color = class))
````

````r
ggplot(data = mpg) + 
geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv))
````

So you see, its very easy to create different plots with the same template. Moreover, **you can also combine different geoms on top of each other**

````r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
geom_point(mapping = aes(color = class)) +
geom_smooth()
````

In the above plot, we are using a global mapping which is applied to both geom_point() and geom_smooth(). However, we can also add individual mappings to the geom functions.

````r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
geom_point(mapping = aes(color = class)) + 
geom_smooth(data = filter(mpg, class == "subcompact"))
````

## Statistical Transformations

We will look at some other different types of graphs - bar charts. Bar charts help us visualise categorical data - data which is in the form of categories. Here we will work the **diamonds** dataset which has columns related to different properties of diamonds - color, carat, price, cut etc...

````r
ggplot(data = diamonds) + 
geom_bar(mapping = aes(x = cut))
````

On the x-axis, the chart displays cut, a variable from diamonds. On the y-axis, it displays count, but count is not a variable in diamonds! Where does count come from? Many graphs, like scatterplots, plot the raw values of your dataset. Other graphs, like bar charts, calculate new values to plot:

1. bar charts, histograms, and frequency polygons bin your data and then plot bin counts, the number of points that fall in each bin
2. smoothers fit a model to your data and then plot predictions from the model.
3. boxplots compute a robust summary of the distribution and then display a specially formatted box.

The algorithm used to calculate new values for a graph is called a **stat, short for statistical transformation**. Each geom function has a default stat() function which can be seen by looking up the documentation for that function. For geom_bar() it is stat_count(). The amazing thing is that we can use the geom() functions and stat() functions intechangeably.

````r
ggplot(data = diamonds) + 
stat_count(mapping = aes(x = cut))
````

We can also override the default stat() for geom_bar(). Instead of counts we can display the proportion of counts

````r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, y = ..prop.., group = 1))
````

You can also use different stat() functions to display different statistical properties of the data

````r
ggplot(data = diamonds) + 
stat_summary(
    mapping = aes(x = cut, y = depth),
    fun.ymin = min,
    fun.ymax = max,
    fun.y = median
)
````

## Position and Color Adjustments

Just like we used different aesthetic mappings in the point() and smooth() functions, we can add different aesthetics in bar plots

````r
ggplot(data = diamonds) + 
geom_bar(mapping = aes(x = cut, colour = cut))
````

````r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = cut))
````

The **fill** attribute is used more often than the **color** attribute. However, these attributes are used to map different variables.

````r
ggplot(data = diamonds) + 
geom_bar(mapping = aes(x = cut, fill = clarity))
````

We see that each cut of the diamond is filled with the clarity attribute. And within each cut we can how the clarity of diamonds is changing. This is one way of looking at the data. There can be multiple ways in which the same data can be visualised. The above type of aesthetic mapping in bar plots is known as **stacking** because different layers are stacked over each other. However, in many scenarios, visualising and comparing proportions is more helpful than comparing counts

````r
ggplot(data = diamonds) + 
geom_bar(mapping = aes(x = cut, fill = clarity), position = "fill")
````

The **position = "fill"** attribute makes the bars equal in height which makes it easy to compare the layers. Each layer height now represents proportion of the classes rather than counts.

````r
ggplot(data = diamonds) + 
geom_bar(mapping = aes(x = cut, fill = clarity), position = "dodge")
````

The previous graph had cut attributes with different clarity values stacked over each other. However, we can also plot them side-by-side which helps in easily visualising the variations of clarity among different cut diamonds. For this we use **position = "dodge"** attribute.

## Coordinate Systems

This is one of the complicated part of using ggplot2. Ggplot2 uses the Cartesian coordinate system as its default - where the X-Axis and Y-Axis determine the position of a point independently. However, there are a number of functions in ggplot2 which help change the coordinate axis and project a different view of data

### 1. coord_flip()

````r
ggplot(data = diamonds)+ 
geom_bar(mapping = aes(x = cut))+
coord_flip()
````

This function simply flips the x and y axis. This is used when there are more variables to analyse on the x-axis and the plots become crowded.

### 2. coord_polar()

````r
bar <- ggplot(data = diamonds) + 
  geom_bar(
    mapping = aes(x = cut, fill = cut), 
    show.legend = FALSE,
    width = 1
  ) + 
  theme(aspect.ratio = 1) +
  labs(x = NULL, y = NULL)

bar + coord_flip()
bar + coord_polar()
````

## Grammar of Graphics

This is the fundamental concept on which ggplot2 is based. The following template is valid for all the graphs/plots which have been plotted using ggplot2

```
ggplot(data = <DATA>) + 
  <GEOM_FUNCTION>(
     mapping = aes(<MAPPINGS>),
     stat = <STAT>, 
     position = <POSITION>
  ) +
  <COORDINATE_FUNCTION> +
  <FACET_FUNCTION>
```

1. Our new template takes seven parameters, the bracketed words that appear in the template. In practice, you rarely need to supply all seven parameters to make a graph because ggplot2 will provide useful defaults for everything except the data, the mappings, and the geom function.

2. **The seven parameters in the template compose the grammar of graphics, a formal system for building plots**. The grammar of graphics is based on the insight that you can uniquely describe any plot as a combination of a dataset, a geom, a set of mappings, a stat, a position adjustment, a coordinate system, and a faceting scheme.

3. You could use this method to build any plot that you imagine. In other words, you can use the code template that you’ve learned in this chapter to build hundreds of thousands of unique plots.
